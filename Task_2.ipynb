```python
import pandas as pd
data=pd.read_csv("/content/ChanDarren_RaiTaran_Lab2a.csv")
data
```

```python
data.info()
```

```python
columns_to_drop = ['PassengerId','Name','Ticket','Cabin']
data = data.drop(columns=[col for col in columns_to_drop if col in data.columns])
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data['Sex']=data['Sex'].map({'male':0,'female':1})
print(data['Sex'].value_counts())
```

```python
data.describe()
data['Sex'].value_counts()
data['Survived'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Survived',data=data)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Sex',hue='Survived', data=data)
plt.title("Survival by gender")
plt.show()
```

```python
sns.countplot(data=data,x='Pclass',hue='Survived')
plt.title('Survival by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=data,x='Age',hue='Survived',multiple='stack', bins=30,kde=False)
plt.title("Survival by age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
data_numeric = data.drop(columns=['Embarked'])
sns.heatmap(data_numeric.corr(),annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
